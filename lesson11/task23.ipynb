### Пример 1

Исполнитель Тренер преобразует число на экране. У исполнителя есть две команды, которым присвоены номера:

1. Прибавить 1
2. Прибавить 2

Первая команда увеличивает число на экране на 1, вторая увеличивает его на 2. Программа для исполнителя Тренер – это последовательность команд.
Сколько существует программ, для которых при исходном числе 1 результатом является число 12?

Заметим, что можно свести задачу к рекурсии. А именно: сколько существует программ для 1->12? Это общее количество программ для 1->11 и 1->10. Из этих двух программ добавлением одной команды (2 и 1 соответственно) получается программа для 1->12.

А теперь сколько существует программ для 1->1? Очевидно, что одна - ничего не делать. Таким образом, получаем:
```
F(n) = 1, при n = 1
F(n) = F(n - 1) + F(n - 2), при n > 1
```

Нужно получить F(12). Такие задачи мы уже умеем решать

````python
F = [0] * 13

for i in range(1, 13):
    if i == 1:
        F[i] = 1
    else:
        F[i] = F[i - 1] + F[i - 2]

print(F[12])
````

### Пример 2

Исполнитель РазДва преобразует число на экране.
У исполнителя есть две команды, которым присвоены номера:

1. Прибавить 1
2. Умножить на 2

Первая команда увеличивает число на экране на 1, вторая умножает его на 2. Программа для исполнителя РазДва — это последовательность команд.

Сколько существует программ, для которых при исходном числе 3 результатом является число 37 и при этом траектория вычислений содержит число 18?

````python
F = [0] * 38

# сначала вычислим все значения до 18 включительно
for i in range(3, 19):
    if i == 3:
        F[i] = 1
    else:
        F[i] = F[i - 1]
        
        # если i четное, то оно могло получиться с помощью команды 2
        if i % 2 == 0:
            F[i] += F[i // 2]

# теперь обнулим все значения кроме 18
for i in range(3, 18):
    F[i] = 0

# теперь, когда мы вычислим остальные значения, то будем учитывать
# только программы через 18 (остальные были обнулены) 
for i in range(19, 38):
    F[i] = F[i - 1]
        
    # если i четное, то оно могло получиться с помощью команды 2
    if i % 2 == 0:
        F[i] += F[i // 2]

print(F[37])
````

### Пример 3

Исполнитель РазДваТри преобразует число на экране. У исполнителя есть три команды, которым присвоены номера:

1. Прибавить 1
2. Умножить на 2
3. Прибавить 3

Первая команда увеличивает число на экране на 1, вторая умножает его на 2, третья увеличивает на 3. 
Программа для исполнителя РазДваТри — это последовательность команд.

Сколько существует программ, которые преобразуют исходное число 2 в число 14 и при этом траектория вычислений не содержит чисел 5 и 10? 

````python
F = [0] * 15

# сначала вычислим все значения до 18 включительно
for i in range(2, 15):
    if i == 2:
        F[i] = 1
    elif i == 5 or i == 10:
        # обнуляем программы, которые проходят через 5 или 10
        F[i] = 0
    else:
        F[i] = F[i - 1] + F[i - 3]
        
        # если i четное, то оно могло получиться с помощью команды 2
        if i % 2 == 0:
            F[i] += F[i // 2]

print(F[14])
````
