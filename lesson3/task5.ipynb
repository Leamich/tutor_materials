### Пример 1

Алгоритм получает на вход натуральное число N ≥ 10 и строит по нему новое число R следующим образом:
1. Все пары соседних цифр в десятичной записи N рассматриваются как двузначные числа (возможно, с ведущим нулём).
2. Из списка полученных на предыдущем шаге двузначных чисел выделяются наименьшее и наибольшее.
3. Результатом работы алгоритма становится сумма найденных на предыдущем шаге двух чисел.

Пример. Дано число N = 2022. Алгоритм работает следующим образом:
1. В десятичной записи выделяем двузначные числа: 20, 02, 22.
2. Наименьшее из найденных чисел 02, наибольшее 22.
3. 02 + 22 = 24.

Результат работы алгоритма R = 24.

При каком наименьшем N в результате работы алгоритма получится R = 137?

```python
def f(N):
    a = []                             # создаём список для хранения получившихся чисел
    
    N = str(N)                         # нам легче работать с N как со строкой
    for i in range(len(N) - 1):        # перебираем все элементы кроме последнего (т.к. для него нет элемента спереди)
        a.append(int(N[i] + N[i + 1])) # формируем число из текущего элемента и следующего
    
    R = max(a) + min(a)                # формируем число, как сказано в условии
    return R


for N in range(10, 1000):
    R = f(N)
    if R == 137:
        print(N)
```

### Пример 2

Автомат получает на вход четырёхзначное натуральное число и строит новое число по следующему алгоритму.

1. Вычисляются суммы первой и второй, второй и третьей, третьей и четвёртой цифр.
2. Из полученных сумм отбрасывается наибольшая (если таких несколько, то одна из них).
3. Остальные суммы записываются в порядке невозрастания.

Пример. Исходное число: 8529. Суммы: 8+5=13;  5+2=7;  2+9=11. Отбрасывается наибольшая сумма 13. Результат: 117.

Укажите наибольшее число, при вводе которого автомат выдаёт значение 127.

```python
def f(N):
    # находим все цифры числа
    a = N // 1000
    b = N % 1000 // 100
    c = N % 100 // 10
    d = N % 10
    
    # можно сделать по-другому
    N = str(N)
    a = int(N[0])
    b = int(N[1])
    c = int(N[2])
    d = int(N[3])
    
    # формируем массив, как в условии
    arr = [
        a + b,
        b + c,
        c + d
    ]
    
    # сортируем его по невозрастанию
    arr = sorted(arr, reverse=True)
    
    # тогда наибольший элемент будет в начале - удалим его
    arr.pop(0)
    
    # формируем число, как сказано в условии
    R = str(arr[0]) + str(arr[1])
    return int(R)


for N in range(1000, 10000):
    R = f(N)
    if R == 127:
        print(N)
```

### Пример 3

Автомат получает на вход трёхзначное число. По этому числу строится новое число по следующим правилам.

1. Перемножаются все цифры исходного числа.
2. Суммируются все цифры исходного числа.
3. Полученные два числа записываются друг за другом в порядке невозрастания (без разделителей).

Пример. Исходное число: 621. Произведение цифр: 6 x 2 x 1 = 12; сумма цифр:  6 + 2 + 1 = 9. Результат: 129.

Укажите наибольшее число, при обработке которого автомат выдаст число 33621.

```python
def f(N):
    N = str(N)
    
    # суммируем все цифры
    s = 0
    for i in range(len(N)):
        s += int(N[i])
        
    # перемножаем все цифры
    m = 1
    for i in range(len(N)):
        m *= int(N[i])
    
    # сортируем его по невозрастанию
    arr = sorted([m, s], reverse=True)
    
    # формируем число, как сказано в условии
    R = str(arr[0]) + str(arr[1])
    return int(R)


for N in range(100, 1000):
    R = f(N)
    if R == 33621:
        print(N)
```

### Пример 4

Автомат обрабатывает натуральное число N по следующему алгоритму.\
1.  Строится двоичная запись числа N.\
2.  Удаляется первая слева единица и все следующие непосредственно за ней нули. Если после этого в числе не остаётся цифр, результат этого действия считается равным нулю.\
3.  Полученное число переводится в десятичную запись.

Пример. Дано число N  =  11. Алгоритм работает следующим образом.

1.  Двоичная запись числа N: 1011.\
2.  Удаляется первая единица и следующий за ней ноль: 11.\
3.  Десятичное значение полученного числа 3.

Сколько разных значений будет показано на экране автомата при последовательном вводе всех натуральных чисел от 10 до 1000?

```python
from collections import Counter


### ШАБЛОН 2.1
def to_q(n, q):
    if n == 0:               
        return "0"
    
    res = ""                
    
    while n:                
        res += str(n % q)   
        n //= q             

    return res[::-1]        
### ШАБЛОН 2.1


def f(N):
    # работаем со списком
    R = list(to_q(N, 2))
    
    # удаляем первый элемент
    R.pop(0)
    
    # удаляем первый элемент, пока он равен 0
    # также проверяем (первое условие), что в списке вообще есть что-то
    while R and R[0] == "0":
        R.pop(0)
    
    # преобразуем список в строку
    R = "".join(R)
    
    # если строка пустая, считаем её как 0
    if not R:
        R = "0"
    
    return int(R)


# записываем все значения в массив
arr = []
for N in range(10, 1000):
    R = f(N)
    arr.append(R)

# считаем все значения массива
cnt = Counter(arr)

# получаем список уникальных значений
values = cnt.keys()
print(len(values)) # выводим его длину
```
